This notebook contains an EDA for the Kaggle creditcard.csv file

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data from the fraud-dedection.csv
df = pd.read_csv("dataset/creditcard.csv")
```

```python
# Basic dataset info
print("Shape:", df.shape)
print("\nData types:")
print(df.dtypes)
```

```python
print("\nFirst rows:")
display(df.head())
```

```python
# Missing values
print("\nMissing values:")
missing = df.isnull().sum()
if missing.sum() == 0:
    print("No missing values found.")
else:
    display(missing[missing > 0])
```

```python
# Class distribution
counts = df['Class'].value_counts()
percentages = df['Class'].value_counts(normalize=True) * 100
class_dist = pd.DataFrame({'Count': counts, 'Percentage': percentages.round(2)})
print("\nClass distribution:")
display(class_dist)

plt.figure(figsize=(5, 4))
plt.bar(counts.index, counts.values)
plt.title("Class distribution (0 = no fraud, 1 = fraud)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
#Statistical summary
print("\nStatistical summary:")
display(df.drop(columns=['Class']).describe().T)
```

```python
# Distribution of numberical features
df.drop(columns=['Class']).hist(figsize=(20, 15), bins=30)
plt.suptitle("Distribution of the numerical features", fontsize=16)
plt.show()
```
